```python
def subtractor(startFileNum, endFileNum):
    test_img_num = startFileNum
    while test_img_num < endFileNum-1:
        testImgDir = VIDEO_DIR+"sample1_"+str(test_img_num)+".png"
        nextImg = VIDEO_DIR+"sample1_"+str(test_img_num+1)+".png"
        img = cv2.imread(testImgDir)
        img2 = cv2.imread(nextImg)
        subtracted_img = img - img2
        exportDir = EXPORT_DIR+"sample1_s"+str(test_img_num)+".png"
        cv2.imwrite(exportDir, subtracted_img)
        test_img_num += 1
```
