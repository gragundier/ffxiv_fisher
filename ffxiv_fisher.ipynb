```python
def subtractor(startFileNum, endFileNum):
    test_img_num = startFileNum
    while test_img_num < endFileNum-1:
        testImgDir = VIDEO_DIR+SAMPLE_NAME+"_"+str(test_img_num)+".png"
        nextImg = VIDEO_DIR+SAMPLE_NAME+"_"+str(test_img_num+1)+".png"
        img = cv2.imread(testImgDir)
        img2 = cv2.imread(nextImg)
        subtracted_img = img2-img
        exportDir = EXPORT_DIR+SAMPLE_NAME+"_s"+str(test_img_num)+".png"
        cv2.imwrite(exportDir, subtracted_img)
        test_img_num += 1
```
